```python
car = { "brand": "Ford",
            "model": "Mustang",
             "year": 1964,
             "year": 2020
             }
car["color"] = "red"
print(type(car))
print(car)
x = car.values()
print(car.values()) #after the change
```

```python
student = {
    "roll": "25070123121",
    "name": "Vedant Trivedi",
    "branch": "ENTC"
}
print("Name:", student["name"])
```

Removing Elements

```python
student.pop("branch")
print("After removals:", student)
```

Updating values

```python
products = {"pen": 18, "Book": 50, "Scale": 60, "Book": 55}
print("Orignal product prices", products)
products["pen"] = 20
print("New product prices", products)
```

```python
marks = {"A": 85, "B": 72, "C": 90}
name = input ("Enter Student name: ")
print(marks.get(name, "student not found"))
```

```python
users = {"admin": "1234", "student": "abcd"}
user_name = input("Username: ")
pw = input("Password: ")
if users.get(user_name) == pw:
    print("Access Granted")
else:
    print("Wrong Password or Username")
```

```python
marks = {"A":85, "B": 92, "C": 78}
topper = max(marks, key=marks.get)
print("Topper: ",topper)
print("Marks: ", marks[topper])
```
